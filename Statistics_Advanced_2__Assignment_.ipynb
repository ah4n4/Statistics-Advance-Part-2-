# 1.What is hypothesis testing in statistics?
  - **Hypothesis testing** in statistics is a method used to make decisions or inferences about a population based on sample data.

---

### ðŸ” **Definition:**

Hypothesis testing is a formal procedure to test assumptions (called **hypotheses**) about a population parameter using sample data.

---

### âœ… **Key Terms:**

* **Null Hypothesis (Hâ‚€):**
  The default or initial assumption (e.g., *There is no difference*, *The mean is equal to a certain value*).

* **Alternative Hypothesis (Hâ‚ or Ha):**
  The statement we want to test for (e.g., *There is a difference*, *The mean is not equal to a certain value*).

* **Test Statistic:**
  A value calculated from sample data that is used to decide whether to reject the null hypothesis.

* **P-value:**
  The probability of observing a test statistic as extreme as, or more extreme than, the value observed, assuming the null hypothesis is true.

* **Significance Level (Î±):**
  A threshold (commonly 0.05) used to decide whether the p-value is small enough to reject the null hypothesis.

---

### ðŸ§ª **Steps in Hypothesis Testing:**

1. **State the hypotheses:**

   * Hâ‚€: Î¼ = Î¼â‚€
   * Hâ‚: Î¼ â‰  Î¼â‚€ (or >, < depending on the test)

2. **Choose a significance level (Î±):**
   Common values: 0.05, 0.01

3. **Collect data and calculate the test statistic.**

4. **Find the p-value or compare with a critical value.**

5. **Make a decision:**

   * If **p-value â‰¤ Î±**, reject Hâ‚€.
   * If **p-value > Î±**, fail to reject Hâ‚€.

---

### ðŸ“˜ **Example:**

Suppose a company claims the average battery life of a phone is 10 hours.

* Hâ‚€: Î¼ = 10
* Hâ‚: Î¼ â‰  10

You test 30 phones and get an average of 9.5 hours. If the p-value is 0.03 and Î± = 0.05, you **reject** the null hypothesis, meaning the battery life is **likely not** 10 hours.

---

### ðŸŽ¯ **Purpose:**

To decide whether there is enough evidence in a sample to support a specific belief about a population.


# 2.What is the null hypothesis, and how does it differ from the alternative hypothesis?
  - ### âœ… What is the **Null Hypothesis** and how does it differ from the **Alternative Hypothesis**?

---

### ðŸ”¹ **1. Null Hypothesis (Hâ‚€):**

* It is the **default assumption** or status quo.
* It states that **there is no effect, no difference, or no relationship** between variables.
* It is what you assume to be true **unless** evidence suggests otherwise.

**Example:**
Hâ‚€: The average height of students is **equal to** 160 cm.

---

### ðŸ”¸ **2. Alternative Hypothesis (Hâ‚ or Ha):**

* It is the **opposite** of the null hypothesis.
* It states that there **is an effect, a difference, or a relationship**.
* It is what you are trying to **provide evidence for**.

**Example:**
Hâ‚: The average height of students is **not equal to** 160 cm.

---

### ðŸ” **Key Differences:**

| Feature               | Null Hypothesis (Hâ‚€)                      | Alternative Hypothesis (Hâ‚ or Ha)  |
| --------------------- | ----------------------------------------- | ---------------------------------- |
| **Assumption**        | No change or effect                       | There is a change or effect        |
| **Symbol**            | Hâ‚€                                        | Hâ‚ or Ha                           |
| **Test Outcome**      | Either "fail to reject Hâ‚€" or "reject Hâ‚€" | Aim to gather evidence against Hâ‚€  |
| **Decision Based On** | Data supports Hâ‚€ unless proven otherwise  | Data supports Hâ‚ if Hâ‚€ is rejected |
| **Role in Testing**   | Starting point                            | Competing claim                    |

---

### ðŸ§ª Example Scenario:

A medicine company claims a drug is effective.

* **Hâ‚€:** The drug has **no effect** (same as a placebo).
* **Hâ‚:** The drug **has an effect** (better or worse than placebo).

If data shows a significant improvement, we **reject Hâ‚€** and accept Hâ‚.

---

### ðŸ“ Summary:

* **Null Hypothesis (Hâ‚€):** No difference or effect â€” status quo.
* **Alternative Hypothesis (Hâ‚):** Indicates a new discovery â€” what you're testing for.
  The goal of hypothesis testing is to determine whether there is **enough evidence** to reject Hâ‚€ in favor of Hâ‚.


# 3.: Explain the significance level in hypothesis testing and its role in deciding the outcome of a test.
  - ### ðŸŽ¯ Significance Level in Hypothesis Testing

---

### ðŸ”¹ **What is the Significance Level (Î±)?**

The **significance level**, denoted by **Î± (alpha)**, is the **threshold probability** used to decide whether to reject the **null hypothesis (Hâ‚€)**.

It represents the **maximum acceptable risk** of making a **Type I error** â€” rejecting Hâ‚€ when it is actually true.

---

### âœ… **Common Values of Î±:**

* **0.05 (5%)** â€” most commonly used
* **0.01 (1%)** â€” for more strict testing
* **0.10 (10%)** â€” for more lenient testing

---

### ðŸ§ª **Role in Hypothesis Testing:**

1. **Set Î± before testing.**
   Example: Î± = 0.05

2. **Calculate the p-value** from the sample data.

3. **Compare p-value with Î±:**

   | p-value | Decision          |
   | ------- | ----------------- |
   | â‰¤ Î±     | Reject Hâ‚€         |
   | > Î±     | Fail to reject Hâ‚€ |

---

### ðŸ“Œ **Example:**

You want to test if a coin is fair.

* **Hâ‚€:** The coin is fair (P = 0.5)
* You get a p-value of 0.03
* If **Î± = 0.05**, then **0.03 < 0.05**, so you **reject Hâ‚€**

This means you have enough evidence to say the coin might not be fair.

---

### âš ï¸ **Why It Matters:**

* A **lower Î±** means you're **less likely** to make a Type I error, but **more likely** to make a Type II error (failing to reject a false Hâ‚€).
* The **choice of Î±** depends on how serious the consequences are for making a wrong decision.

---

### ðŸ“˜ **Summary:**

* The **significance level (Î±)** sets the **cutoff for decision-making**.
* It helps control the risk of **false positives**.
* You **reject Hâ‚€** if your p-value is **less than or equal to Î±**.


# 4.What are Type I and Type II errors? Give examples of each.
  - ### âš–ï¸ Type I and Type II Errors in Hypothesis Testing

When making decisions based on hypothesis testing, two kinds of errors can occur:

---

### ðŸ”´ **1. Type I Error (False Positive):**

* **Definition:** Rejecting the null hypothesis (**Hâ‚€**) when it is actually **true**.
* **Symbol:** Î± (alpha) â€” the **significance level** is the probability of making a Type I error.
* **Think of it as:** Detecting an effect that **isn't really there**.

**ðŸ“Œ Example:**

* **Hâ‚€:** A person is **not guilty**.
* **Hâ‚:** A person is **guilty**.
* A **Type I error** occurs if the court **convicts an innocent person**.

---

### ðŸ”µ **2. Type II Error (False Negative):**

* **Definition:** Failing to reject the null hypothesis (**Hâ‚€**) when it is actually **false**.
* **Symbol:** Î² (beta) â€” the probability of making a Type II error.
* **Think of it as:** **Missing a real effect** that actually exists.

**ðŸ“Œ Example:**

* **Hâ‚€:** A person is **not guilty**.
* **Hâ‚:** A person is **guilty**.
* A **Type II error** occurs if the court **lets a guilty person go free**.

---

### ðŸ” **Comparison Table:**

| Error Type  | What Happens                          | Reality         | Decision Made  | Risk Symbol |
| ----------- | ------------------------------------- | --------------- | -------------- | ----------- |
| **Type I**  | Reject Hâ‚€ when it's **true**          | Hâ‚€ is **true**  | False positive | Î±           |
| **Type II** | Fail to reject Hâ‚€ when it's **false** | Hâ‚€ is **false** | False negative | Î²           |

---

### ðŸ§ª Real-Life Example: Medical Test

* **Hâ‚€:** Patient does **not** have the disease.
* **Hâ‚:** Patient **has** the disease.

| Situation                        | Type of Error                      |
| -------------------------------- | ---------------------------------- |
| Healthy person diagnosed as sick | **Type I Error** (False positive)  |
| Sick person diagnosed as healthy | **Type II Error** (False negative) |

---

### âœ… Summary:

* **Type I Error (Î±):** Rejecting a true Hâ‚€ â€” a **false alarm**.
* **Type II Error (Î²):** Not rejecting a false Hâ‚€ â€” **missing a real issue**.

Both errors are important, and the choice of significance level (Î±) affects the balance between them.


# 5. What is the difference between a Z-test and a T-test? Explain when to use each.
   - ### ðŸ“Š Difference Between Z-test and T-test

Both **Z-test** and **T-test** are statistical methods used to test hypotheses about population means, but they are used in **different situations** based on sample size and known parameters.

---

### ðŸ”¹ **1. Z-Test**

#### âœ… When to Use:

* **Large sample size** (n â‰¥ 30)
* **Population standard deviation (Ïƒ)** is **known**
* Data is approximately normally distributed

#### ðŸ“Œ Example:

Testing if the average weight of a population is 70 kg using a sample of 100 people and knowing the population standard deviation.

---

### ðŸ”¸ **2. T-Test**

#### âœ… When to Use:

* **Small sample size** (n < 30)
* **Population standard deviation is unknown**
* Data should be approximately **normally distributed**

#### ðŸ“Œ Example:

Testing if a new teaching method affects average test scores using a sample of 15 students, and the population standard deviation is **not known**.

---

### ðŸ” **Key Differences:**

| Feature                  | Z-Test                         | T-Test                              |
| ------------------------ | ------------------------------ | ----------------------------------- |
| **Sample Size**          | Large (n â‰¥ 30)                 | Small (n < 30)                      |
| **Population Std. Dev.** | Known                          | Unknown                             |
| **Test Statistic**       | Z-value (from standard normal) | t-value (from t-distribution)       |
| **Distribution Used**    | Standard Normal (Z)            | Studentâ€™s t-distribution            |
| **More Conservative?**   | No                             | Yes (accounts for more uncertainty) |

---

### ðŸ“˜ Types of T-tests:

* **One-sample T-test:** Compare sample mean to a known value
* **Independent two-sample T-test:** Compare means of two independent groups
* **Paired T-test:** Compare means of the same group at different times

---

### âœ… Summary:

\| Use a **Z-test** when:
â†’ You have a **large sample** and know the **population standard deviation**.

\| Use a **T-test** when:
â†’ You have a **small sample** and do **not know** the population standard deviation.

```python
# 7. Implement hypothesis testing using Z-statistics for a sample dataset in Python. Show the Python code and interpret the results. sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,50.3, 50.4, 50.0, 49.7, 50.5, 49.9
import numpy as np
from scipy.stats import norm

# Sample data
sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
               50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
               50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
               50.3, 50.4, 50.0, 49.7, 50.5, 49.9]

# Hypothesized population mean
mu_0 = 50.0

# Calculate sample statistics
sample_mean = np.mean(sample_data)
sample_std = np.std(sample_data, ddof=1)  # sample standard deviation
n = len(sample_data)

# Calculate Z-statistic
z_stat = (sample_mean - mu_0) / (sample_std / np.sqrt(n))

# Calculate two-tailed p-value
p_value = 2 * (1 - norm.cdf(abs(z_stat)))

# Print results
print(f"Sample Mean = {sample_mean:.4f}")
print(f"Sample Standard Deviation = {sample_std:.4f}")
print(f"Z-Statistic = {z_stat:.4f}")
print(f"P-Value = {p_value:.4f}")

# Interpretation
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: There is a significant difference from 50.0.")
else:
    print("Fail to reject the null hypothesis: No significant difference from 50.0.")
```

```python
# 8.Write a Python script to simulate data from a normal distribution and calculate the 95% confidence interval for its mean. Plot the data using Matplotlib.
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# Step 1: Simulate data
np.random.seed(42)  # for reproducibility
mu = 100     # true mean
sigma = 15   # standard deviation
n = 100      # sample size

data = np.random.normal(mu, sigma, n)

# Step 2: Calculate 95% confidence interval for the mean
sample_mean = np.mean(data)
sample_std = np.std(data, ddof=1)
confidence_level = 0.95
z_critical = stats.norm.ppf(1 - (1 - confidence_level) / 2)  # two-tailed z-value

margin_of_error = z_critical * (sample_std / np.sqrt(n))
ci_lower = sample_mean - margin_of_error
ci_upper = sample_mean + margin_of_error

print(f"Sample Mean = {sample_mean:.2f}")
print(f"95% Confidence Interval = ({ci_lower:.2f}, {ci_upper:.2f})")

# Step 3: Plot the data and confidence interval
plt.figure(figsize=(10, 6))
plt.hist(data, bins=15, color='skyblue', edgecolor='black', alpha=0.7)
plt.axvline(sample_mean, color='red', linestyle='--', label='Sample Mean')
plt.axvline(ci_lower, color='green', linestyle='--', label='95% CI Lower')
plt.axvline(ci_upper, color='green', linestyle='--', label='95% CI Upper')
plt.title('Histogram of Normally Distributed Data with 95% Confidence Interval')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.show()

```

```python
# 9.Write a Python function to calculate the Z-scores from a dataset and visualize the standardized data using a histogram. Explain what the Z-scores represent in terms of standard deviations from the mean.
import numpy as np
import matplotlib.pyplot as plt

def calculate_and_plot_z_scores(data):
    # Calculate mean and standard deviation
    mean = np.mean(data)
    std = np.std(data)

    # Calculate Z-scores
    z_scores = (data - mean) / std

    # Print first 5 Z-scores
    print("First 5 Z-scores:", z_scores[:5])

    # Plot histogram
    plt.figure(figsize=(10, 6))
    plt.hist(z_scores, bins=15, color='orange', edgecolor='black', alpha=0.7)
    plt.title('Histogram of Z-scores (Standardized Data)')
    plt.xlabel('Z-score')
    plt.ylabel('Frequency')
    plt.grid(axis='y', linestyle='--', alpha=0.6)
    plt.axvline(0, color='red', linestyle='--', label='Mean (Z=0)')
    plt.legend()
    plt.show()

    return z_scores

# Example dataset
data = np.random.normal(loc=100, scale=15, size=100)

# Call the function
z_scores = calculate_and_plot_z_scores(data)
```
